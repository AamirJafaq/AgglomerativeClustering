```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python
df=pd.read_csv("shopping-data.csv")
df.head()
```

```python
print("Consider the last two columns")
data=df.iloc[:, [3,4]]
data.head()
```

```python
sns.relplot(data, x="Annual Income (k$)", y="Spending Score (1-100)", height=3)
```

```python
sp.cluster.hierarchy.dendrogram(sp.cluster.hierarchy.linkage(data, method='ward'))
plt.show()
```

# AgglomerativeClustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agglo=AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
agglo=agglo.fit(data)
```

```python
plt.figure(figsize=(10, 7))
sns.relplot(x=data.iloc[:,0], y=data.iloc[:,1], hue=agglo.labels_, kind="scatter", height=3)
plt.show()
```
